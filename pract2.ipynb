```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import accuracy_score

X = np.linspace(0, 10, 50).reshape(-1, 1)
y = 3 * X.flatten() + 2 + np.random.randn(50) * 3

lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_lin = lin_reg.predict(X)

poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
y_pred_poly = poly_reg.predict(X_poly)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X, y, color="blue", label="Data")
plt.plot(X, y_pred_lin, color="red", label="Linear Fit")
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X, y, color="blue", label="Data")
plt.plot(X, y_pred_poly, color="green", label="Polynomial Fit (deg=3)")
plt.title("Polynomial Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()

plt.tight_layout()
plt.show()

np.random.seed(42)
X_class = np.linspace(-10, 10, 100).reshape(-1, 1)
y_class = (X_class.flatten() > 0).astype(int)
y_class = np.where(np.random.rand(100) > 0.9, 1 - y_class, y_class)

log_reg = LogisticRegression()
log_reg.fit(X_class, y_class)
y_prob = log_reg.predict_proba(X_class)[:, 1]

plt.figure(figsize=(6, 5))
plt.scatter(X_class, y_class, color="blue", label="Data", alpha=0.6)
plt.plot(X_class, y_prob, color="red", linewidth=2, label="Logistic Curve")
plt.title("Logistic Regression")
plt.xlabel("X")
plt.ylabel("Probability of Class 1")
plt.legend()
plt.show()

y_pred_class = log_reg.predict(X_class)
print("Logistic Regression Accuracy:", accuracy_score(y_class, y_pred_class))
```
