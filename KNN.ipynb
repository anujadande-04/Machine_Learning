```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import math

df = pd.read_excel(r"C:\Users\Anuja\OneDrive\Documents\ML_Practicals\Coffee.xlsx")
print(df.head())

caffeine = df["Caffeine_mg"].values
serving = df["Serving_ml"].values
labels = df["Type"].values

x1 = 110   
y1 = 230   

distances = []
for i in range(len(caffeine)):
    d = math.sqrt((x1 - caffeine[i])**2 + (y1 - serving[i])**2)
    distances.append(d)
    print(f"Distance from {labels[i]}: {d}")

nearest_index = np.argmin(distances)

print("\nNearest distance:", distances[nearest_index])
print("Nearest neighbor:", (caffeine[nearest_index], serving[nearest_index]))
print("Predicted Coffee Type:", labels[nearest_index])

plt.scatter(caffeine, serving, c="blue", label="Existing Coffees")
plt.scatter(x1, y1, c="red", marker="x", s=100, label="New Coffee")
plt.xlabel("Caffeine (mg)")
plt.ylabel("Serving Size (ml)")
plt.title("KNN Coffee Classification (k=1)")
plt.legend()
plt.show()
```
